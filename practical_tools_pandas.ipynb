função criada para encontrar a posição de determinado elemento no dataframe

```python
def encontrar_posicao(valor_procurado, dataframe):
    for coluna in dataframe.columns:
        if valor_procurado in dataframe[coluna].values:
            linha = dataframe.index[dataframe[coluna] == valor_procurado][0]
            return [linha, coluna]

    return print(f'O dataframe não contém o {valor_procurado}')
```

A função percorre cada valor na lista e, para cada valor, verifica se ele está presente em alguma coluna do DataFrame. Se o valor for encontrado, a função imprime a posição [linha, coluna] onde o valor está localizado.

```python
def find_position_in_all_column(lista, dataframe):
    for i, val in enumerate(lista):
        for j, coluna in enumerate(dataframe.columns):
            if val in dataframe[coluna].values:
                linha = dataframe.index[dataframe[coluna] == val][0]
                print(f'O valor {val} está na posição [{linha}, {coluna}].')
```

```python
def bin_to_hex(bin_str):
    # Verifica se a string binária tem 6 dígitos
    if len(bin_str) != 6 or not all(bit in '01' for bit in bin_str):
        raise ValueError("A string binária deve ter 6 dígitos.")

    # Converte a string binária para inteiro decimal
    decimal_num = int(bin_str, 2)

    # Converte o inteiro decimal para hexadecimal
    hex_num = hex(decimal_num)

    # Retorna o resultado removendo o prefixo '0x'
    return hex_num[2:]

# Exemplo de uso
numero_binario = "110110"
numero_hexadecimal = bin_to_hex(numero_binario)
print(f"O número binário {numero_binario} em hexadecimal é: {numero_hexadecimal}")
```

```python
def bin_to_decimal(bin_str):
    # Verifica se a string binária tem 6 dígitos
    if len(bin_str) != 6 or not all(bit in '01' for bit in bin_str):
        raise ValueError("A string binária deve ter 6 dígitos.")

    # Converte a string binária para inteiro decimal
    decimal_num = int(bin_str, 2)

    # Retorna o inteiro decimal
    return decimal_num

# Exemplo de uso
numero_binario = "100111"
numero_decimal = bin_to_decimal(numero_binario)

print(f"O número binário {numero_binario} em decimal é: {numero_decimal}")
```
